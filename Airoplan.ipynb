```python
dataframe_train["Days_Since_Inspection"]=dataframe_train["Days_Since_Inspection"].astype(np.float16)
dataframe_train["Violations"]=dataframe_train["Violations"].astype(np.float16)
dataframe_train["Accident_Type_Code"]=dataframe_train["Accident_Type_Code"].astype(np.float16)
dataframe_train["Total_Safety_Complaints"]=dataframe_train["Total_Safety_Complaints"].astype(np.float16)
```

```python
train_data=dataframe_train.iloc[:,1:11]
```

```python
labels=dataframe_train.iloc[:,0]
```

```python
labels
```

```python
from sklearn.cluster import k_means
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras import optimizers
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import os
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import losses
from tensorflow.keras import metrics
```

```python
encoded=LabelEncoder()
encoded.fit(labels)
labels_encoded=encoded.transform(labels)
```

```python
def build_model():
    model=Sequential()
    model.add(Dense(128,activation="relu",input_shape=(train_data.shape[1],)))
    model.add(Dense(128,activation='relu'))
    model.add(Dense(32,activation='relu'))
    model.add(Dense(4,activation='softmax'))
    
    
    return model
```

```python
trainx,testx,trainy,testy=train_test_split(train_data,labels_encoded,test_size=0.25,random_state=42)
```

```python
model=build_model()
model.compile(loss='sparse_categorical_crossentropy',optimizer="rmsprop",metrics=['accuracy'])
model.fit(trainx,trainy,epochs=20,batch_size=1,validation_data=(testx, testy),verbose=1)
```
