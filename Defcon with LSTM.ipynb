```python
labels=dataframe_train.iloc[:,10]
```

```python
encoded=LabelEncoder()
encoded.fit(labels)
labels_encoded=encoded.transform(labels)
labels_encoded
```

```python
np.unique(labels_encoded)
```

```python
train=train_data.iloc[:7500].values

test=train_data.iloc[7500:].values

label_train=labels_encoded[:7500]

label_test=labels_encoded[7500:]

```

```python
def build_model():
    regressior = Sequential()

    regressior.add(LSTM(units = 60, activation = 'relu', return_sequences = True, input_shape = (110, 10)))
    regressior.add(Dropout(0.2))

    regressior.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
    regressior.add(Dropout(0.2))

    regressior.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
    regressior.add(Dropout(0.2))

    regressior.add(LSTM(units = 120, activation = 'relu'))
    regressior.add(Dropout(0.2))

    regressior.add(Dense(units = 5,activation='softmax'))
    
    
    return regressior
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
#trainx,testx,trainy,testy=train_test_split(train_data,labels_encoded,test_size=0.25,random_state=42)
length=110
generator=TimeseriesGenerator(train,label_train,length=length,batch_size=1,shuffle=True)
```

```python

val_generator=TimeseriesGenerator(test,label_test,length=length,batch_size=1,shuffle=True)
```

```python
from tensorflow.keras.layers import Dropout
```

```python

model=build_model()
model.compile(loss='sparse_categorical_crossentropy',optimizer=Adam(lr=1e-4),metrics=['accuracy'])
#model.fit(trainx,trainy,epochs=10,validation_data=(testx,testy),batch_size=1)
model.fit_generator(generator,epochs=5,validation_data=(val_generator))
```
